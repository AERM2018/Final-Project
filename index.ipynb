```
#Before to run this code execute above git clone 
import time
import random

#initialize variables
num_questions_to_answer = 3
num_questions = 30
decimals = 5

#Print the instructions how to play the game
def show_instructions():
    print("Hello, welcome to the game \"Who is more smart?\"")
    print("""
The instructions are:
1) I´m going to give you a question.
2) The que2stion has 4 posible answers, you've got to select the answer that you think is the correct one
Oh! I forgot to tell you something, each player has 10 seconds to response each question. You've got to be fast!
3) I'm going to tell you the correct answer and after that I´m going to show you the score of each player""")


#get the numbers of players
#Create and initialize the vector which is saving the score of each player    
def get_nplayers():
    num =int(input("\nCould you tell me the numbers of players: "))

    players =  list()
    for i in range(num):
        info_player = {"name":"player "+str(i+1),"score":0,"time":0.0}
        players.append(info_player)
    return num,players

#create a vector and select 10 numbers in a range of 1 and 10
#These numbers are the numbers of question the players are going to answer
def get_randoms():
    random_nums = list()
    while len(random_nums)<num_questions_to_answer:
        r = random.randint(1,num_questions)
        if not r in random_nums:
            random_nums.append(r)
    return random_nums

#get and create a list with the answers of the question selected, shuffle them.
#With the answers shuffuled create a new dictionary and return it
def get_ans_rdm(num_q):
  global q
  pa = list()
  for key in ('a','b','c','d'):
    pa.append(q[num_q][key])
  random.shuffle(pa)
  return dict(zip('abcd',pa))

#Print the score at the moment
def show_score(flag_end):
    global score
    print("\n----------Score---------")
    for s in range(nPlayers):
        print("{}: {} pts".format(score[s]['name'],score[s]['score']))
        if flag_end:
          print("Time: {} sec(s)".format(score[s]['time']))
    print("-------------------------\n")

#sort the score table in order to know who is the firts place depending in the score and the time
#sort by bubble sort method
def sort_score_table():
  global score
  if len(score) > 1:
    for i in range(len(score)):
      for k in range (len(score)-1):
        if score[k+1]['score'] == score[k]['score']:
            if score[k+1]['time'] < score[k]['time'] :
              score[k+1],score[k] = score[k],score[k+1]
        elif score[k+1]['score'] > score[k]['score']:
            score[k+1],score[k] = score[k],score[k+1]


#Start the game
show_instructions()
nPlayers,score = get_nplayers()
rdms = get_randoms() 
for i in range(len(rdms)):
    
    #print the question selected
    print("Question #",i+1)
    print(q[rdms[i]]['text'])

    #get the dictionary with the answers shuffled
    #print the answers
    ans_rdm = get_ans_rdm(rdms[i])
    for a in ans_rdm.keys():
      print(a+") "+ans_rdm[a])
    print("\n")

    #start the timer and wait the user's answer
    for k in range(nPlayers):
        #create a new thread
        t = tt()
        t.setDaemon(True)
        print("\nplayer "+str(k+1))
        #take the time when the  question is showed
        ti = time.time()
        t.start()
        ans = input()
        #if the thread is alive that means the user answered the question in time
        #if the thread is not alive, the user can type any answer but it won't count and add 10 seg in the variable time
        if t.isAlive():
          #take the time when the user enter the answer in time
          tf = time.time()
          #the total time  the user take to answer the question
          totalTime= tf-ti
          t.stop()
          #round the time
          totalTime = round(totalTime,decimals)
          try:
            if ans_rdm[ans] == q[rdms[i]]['correct']:
              score[k]['score'] +=100
              score[k]['time'] +=totalTime
          except :
            print("Invalid answer")
        else: score[k]['time'] +=10
    #show the score without time
    show_score(False)
#show the score with time
show_score(True)
sort_score_table()
print("The winer is: {} with {} points  in {} sec(s)".format(score[0]['name'],score[0]['score'],score[0]['time']))



      
       
        
        

        
```
